```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
);
```

# Get observable-jupyter

```python
!pip install observable-jupyter
from observable_jupyter import embed
```

```python
# Check with version of Plot came across with the API (need 0.3 to get color legend)
embed("@pbogden/observable-plot-jupyter", cells=["version"])
```

# Data

Convert the Pandas dataframe to a (serializable) Python dictionary.

```python
data = tips.to_dict('records')
```

# Dot plot

```python
dot_options = {"x": "total_bill", "y": "tip", "stroke": "none", "fill": "smoker", "r": "size"}

embed("@pbogden/observable-plot-jupyter", cells=["dot_plot"], inputs={"data": data, "dot_options": dot_options})
```

# Data table

```python
embed("@pbogden/observable-plot-jupyter", cells=["data_table"], inputs={"data": data})
```

# Rect plot

```python
bin_outputs = {"fillOpacity": "count"}

bin_options = {"x": "total_bill", "y": "tip", "fill": "smoker"}

embed("@pbogden/observable-plot-jupyter", cells=["rect_plot"], inputs={"data": data, "bin_outputs": bin_outputs, "bin_options": bin_options})
```

# Recty plot

```python
binx_outputs = {"y": "count"}

binx_options = {"x": "total_bill", "fill": "smoker"}

embed("@pbogden/observable-plot-jupyter", cells=["recty_plot"], inputs={"data": data, "binx_outputs": binx_outputs, "binx_options": binx_options})
```

# Facet plot

```python
facet_options = {"data": data, "y": "smoker" }

embed("@pbogden/observable-plot-jupyter", 
      cells=["facet_plot"], 
      inputs={"data": data, "binx_outputs": binx_outputs, "binx_options": binx_options, "facet_options": facet_options})
```
